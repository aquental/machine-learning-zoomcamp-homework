```python
df_asia = df[df['origin'] == 'Asia']
df_asia
```

```python
df_asia = df_asia[['vehicle_weight', 'model_year']]
df_asia
```

```python
df_asia = df_asia.head(7)
df_asia
```

```python
X = df_asia.values
X
```

```python
X.T
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
y
```

```python
w = XTX_inv.dot(X.T)
w
```

```python
w = w.dot(y)
w
```

```python
result = w.sum()
result
```
